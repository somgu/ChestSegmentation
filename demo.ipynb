```python
from keras_unet.models import custom_unet
from keras_unet.utils import plot_imgs
from Dataset import Dataset
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
from Dataset import Dataset

dataset = Dataset("./ChestCT_GT")
dcm_path_list = dataset.dcm_path_list
mask_path_list = dataset.mask_path_list
dcm_list = dataset.dicoms
mask_list = dataset.masks
dcms = []
masks = []
```

```python
for dcm, mask in zip(dcm_list, mask_list):
    dcms.append(dcm.pixel_array)
    masks.append(mask)
```

```python
dcms_np = np.asarray(dcms)
masks_np = np.asarray(masks)
```

```python
plot_imgs(org_imgs=dcms_np, mask_imgs=masks_np, nm_img_to_plot=10, figsize=6)
```
